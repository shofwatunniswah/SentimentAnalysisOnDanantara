```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from datetime import datetime
import matplotlib.pyplot as plt

# Load data hasil scraping
df= pd.read_csv('/content/drive/My Drive/PBA/Danantara_Preprocessed.csv')
df
```

```python
# Ubah kolom tanggal ke datetime, handle error dan format yang berbeda
# Menggunakan format yang sesuai dengan data Anda
# dan errors='coerce' untuk mengubah data yang tidak valid menjadi NaT
df['published_date'] = pd.to_datetime(df['published_date'], errors='coerce')

# Tambah kolom waktu turunan
df['month'] = df['published_date'].dt.month_name()
df['quarter'] = df['published_date'].dt.to_period("Q")
```

```python
df
```

## Fungsi: Top Words berdasarkan TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def top_tfidf_words_all(df, text_col='stemmed', top_n=10):
    vectorizer = TfidfVectorizer(stop_words='english', max_df=0.8)
    tfidf_matrix = vectorizer.fit_transform(df[text_col])
    feature_names = vectorizer.get_feature_names_out()
    mean_scores = tfidf_matrix.mean(axis=0).A1
    top_indices = mean_scores.argsort()[::-1][:top_n]
    top_words = [(feature_names[i], mean_scores[i]) for i in top_indices]
    return top_words

top_words = top_tfidf_words_all(df)
print("Top TF-IDF Words:")
for word, score in top_words:
    print(f"  {word}: {score:.4f}")
```

## Fungsi: Top Words berdasarkan TF-IDF per Grup

```python
def top_tfidf_words(df, group_col, text_col='stemmed', top_n=10):
    results = {}
    for group in df[group_col].dropna().unique():
        subset = df[df[group_col] == group]
        vectorizer = TfidfVectorizer(stop_words='english', max_df=0.8)
        tfidf_matrix = vectorizer.fit_transform(subset[text_col])
        feature_names = vectorizer.get_feature_names_out()
        mean_scores = tfidf_matrix.mean(axis=0).A1
        top_indices = mean_scores.argsort()[::-1][:top_n]
        top_words = [(feature_names[i], mean_scores[i]) for i in top_indices]
        results[str(group)] = top_words
    return results
```

## Jalankan Analisis TF-IDF Berdasarkan Kuartal

```python
top_words_by_quarter = top_tfidf_words(df, group_col='quarter')
for quarter, words in top_words_by_quarter.items():
    print(f"\nTop TF-IDF Words in Quarter {quarter}:")
    for word, score in words:
        print(f"  {word}: {score:.4f}")
```

## Jalankan Analisis TF-IDF Berdasarkan Bulan

```python
top_words_by_month = top_tfidf_words(df, group_col='month')
for month, words in top_words_by_month.items():
    print(f"\nTop TF-IDF Words in {month}:")
    for word, score in words:
        print(f"  {word}: {score:.4f}")
```

## ðŸ“Š Visualisasi TF-IDF Top Words dalam Treemap

```python
import plotly.express as px

# Pilih grup yang ingin divisualisasikan, misal: salah satu kuartal
selected_group = '2025Q1'  # ganti sesuai hasil group yang tersedia
top_words = top_words_by_quarter[selected_group]

# Buat DataFrame untuk visualisasi
treemap_df = pd.DataFrame(top_words, columns=["word", "tfidf_score"])
fig = px.treemap(treemap_df, path=["word"], values="tfidf_score",
                 title=f"Treemap Top TF-IDF Words - {selected_group}",
                 color="tfidf_score", color_continuous_scale="Blues")
fig.show()
```
