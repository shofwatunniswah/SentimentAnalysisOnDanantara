## Deklarasi URL Berita

Pada proses ini saya memasukkan list url berita, yang mana saya menggunakan 3 link berita dan disimpan pada list variable list_of_urls

```python
import pandas as pd

df = pd.read_csv('/content/drive/My Drive/PBA/Link_Scraping_Danantara.csv')
df

```

```python
list_of_urls = df['link'].tolist()
list_of_urls
```

Selanjutnya ini merupakan proses yang dilakukan newspaper3k yaitu scraping data dari web berita, dan menhasilkan output berupa judul artikel, penulis artikel, tanggal publikasi artikel, dan ringkasan artikel. Dan untuk hasil ringkasan dan teks berita disimpan dalam list kosong yang sebelum proses looping sudah di deklarasikan baru di appand pada saar looping.

```python
import newspaper
from newspaper import Config
import pandas as pd
from urllib.parse import urlparse


config = Config()
config.request_timeout = 30
article_details_list = []

# Loop through each URL
for url in list_of_urls:
    try:
        article = newspaper.Article(url=url, language='id', config=config)  # Specify language as 'id' for Indonesian
        article.download()
        article.parse()

        # Attempt to get a summary using article.summary first
        summary = article.summary
        text = article.text
        parsed_url = urlparse(url)
        source = parsed_url.netloc.replace('www.', '')


        # If article.summary is empty, try using meta_data['description']
        if not summary:
            meta_data = article.meta_data
            summary = meta_data.get('description', None)


        article_detail = {
            "title": str(article.title),
            "authors": ", ".join(article.authors) if article.authors else None,  # Handle potential empty author lists
            "source": source,
            "published_date": str(article.publish_date),
            "summary": summary,
            "content": text,
            "url": url,
        }
        article_details_list.append(article_detail)

    except newspaper.ArticleException as e:
        print(f"Failed to process URL: {url}. Error: {e}")
    except Exception as e:  # Catch other potential exceptions during processing
        print(f"An unexpected error occurred for URL {url}: {e}")

df_star = pd.DataFrame(article_details_list)
df_star
```
