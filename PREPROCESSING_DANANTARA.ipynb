```python
import pandas as pd
from textblob import TextBlob
from googletrans import Translator

# Inisialisasi translator
translator = Translator()

# Fungsi untuk menerjemahkan dan menganalisis sentimen
def analyze_sentiment(text):
    try:
        translated = translator.translate(text, src='id', dest='en')
        blob = TextBlob(translated.text)
        return pd.Series([blob.sentiment.polarity, blob.sentiment.subjectivity])
    except Exception as e:
        print(f"Terjadi error: {e}")
        return pd.Series([None, None])

# Terapkan ke DataFrame
df_star[['polarity', 'subjectivity']] = df_star['content'].apply(analyze_sentiment)

# Tampilkan hasil
print(df_star[['content', 'polarity', 'subjectivity']].head())
```

```python
from collections import Counter
import matplotlib.pyplot as plt # Import matplotlib.pyplot
import seaborn as sns # Import seaborn

def freq_words(x, terms=30):
    all_words = ' '.join([text for text in x])
    all_words = all_words.split()

    # Count word frequency using Counter (faster than FreqDist)
    word_counts = Counter(all_words)
    words_df = pd.DataFrame(word_counts.items(), columns=['word', 'count'])

    # Select top 'terms' most frequent words
    d = words_df.nlargest(columns="count", n=terms)

    plt.figure(figsize=(20,5))
    ax = sns.barplot(data=d, x="word", y="count", palette="rainbow")
    ax.set(ylabel='Count')
    plt.show()
```

```python
from textblob import TextBlob
from wordcloud import WordCloud
import pandas as pd
import numpy as np
import re
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import nltk

# Pastikan stopwords dan tokenizer sudah diunduh
nltk.download('stopwords')
nltk.download('punkt')

def normalize_date(date_str):
    try:
        dt = pd.to_datetime(date_str, utc=True)
        return dt.tz_convert(None).strftime('%Y-%m-%d %H:%M:%S')
    except Exception:
        return None

class Preprocess:
    def __init__(self):
        self.stop_words_add = ['baca','com', 'v', 'b','nadia', 'pt', 'penny', 'casino', 'jakarta', 'kompas', 'cansino']
        self.stop_words_id = set(stopwords.words('indonesian'))
        self.stop_words_en = set(stopwords.words('english'))
        self.combined_stopwords = self.stop_words_id.union(self.stop_words_en)
        factory = StemmerFactory()
        self.stemmer = factory.create_stemmer()

    def low(self, data):
        return data.lower()

    def del_num(self, data):
        return re.sub(r"\d", "", data)

    def del_punct(self, data):
        data = re.sub(r'#','', data)
        data = re.sub(r'@[A-Za-z0-9]+', '', data)
        data = re.sub(r'rt[\s]+','', data)
        data = re.sub(r'https?:\/\/\S+', '', data)
        data = re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", data)
        return data

    def del_ws(self, data):
        return data.strip()

    def remove_stopwords_custom(self, tokens):
        return [w for w in tokens if w.lower() not in self.stop_words_add]

    def remove_stopwords_combined(self, tokens):
        return [w for w in tokens if w.lower() not in self.combined_stopwords]

    def apply_stemming(self, tokens):
        return [self.stemmer.stem(w) for w in tokens]

    def proses_semua(self, data):
        result = {}
        data = self.low(data)
        data = self.del_num(data)
        data = self.del_punct(data)
        data = self.del_ws(data)

        tokens_awal = word_tokenize(data)
        tokens_no_stop1 = self.remove_stopwords_custom(tokens_awal)
        tokens_no_stop2 = self.remove_stopwords_combined(tokens_no_stop1)
        tokens_stemmed = self.apply_stemming(tokens_no_stop2)

        result['original'] = data
        result['tokens_awal'] = tokens_awal
        result['no_stopwords'] = tokens_no_stop2
        result['stemmed'] = tokens_stemmed

        return result
```

```python
# Terapkan fungsi ke setiap baris pada kolom 'content'
pp = Preprocess() # Create an instance of the Preprocess class

df_star['published_date'] = df_star['published_date'].apply(normalize_date)

# Convert 'content' column to string type before applying the function
df_star['content'] = df_star['content'].astype(str)

hasil_df = df_star['content'].apply(pp.proses_semua)

# Ubah hasil nested dict menjadi DataFrame
hasil_df = pd.DataFrame(hasil_df.tolist())

# Tampilkan hasil
print(hasil_df.head())
```

## Sentiment analysis


```python
from textblob import TextBlob
from googletrans import Translator
import pandas as pd

translator = Translator()

# Fungsi analisis sentimen dari list token
def analyze_sentiment_from_tokens(tokens):
    try:
        # Pastikan input adalah list
        if not isinstance(tokens, list):
            return pd.Series([None, None])

        # Gabungkan list token jadi kalimat
        text = ' '.join(tokens)

        # Terjemahkan ke Bahasa Inggris
        translated = translator.translate(text, src='id', dest='en')

        # Analisis sentimen
        blob = TextBlob(translated.text)
        return pd.Series([blob.sentiment.polarity, blob.sentiment.subjectivity])

    except Exception as e:
        print(f"Terjadi error: {e}")
        return pd.Series([None, None])

# Terapkan ke kolom 'stemmed'
gabungan_df[['polarity_stemmed', 'subjectivity_stemmed']] = gabungan_df['stemmed'].apply(analyze_sentiment_from_tokens)

# Lihat hasil
print(gabungan_df[['stemmed', 'polarity_stemmed', 'subjectivity_stemmed']].head())
```

### Berdasarkan portal berita

```python
# Fungsi klasifikasi gabungan polaritas dan subjektivitas
def klasifikasi_gabungan(polaritas, subjektivitas):
    if polaritas <= -0.5 and subjektivitas <= 0.5:
        return 'Kritik Objektif'
    elif polaritas <= -0.5 and subjektivitas > 0.5:
        return 'Kritik Subjektif'
    elif -0.1 <= polaritas <= 0.1 and subjektivitas <= 0.5:
        return 'Netral Objektif'
    elif -0.1 <= polaritas <= 0.1 and subjektivitas > 0.5:
        return 'Netral Subjektif'
    elif polaritas >= 0.5 and subjektivitas <= 0.5:
        return 'Pujian Objektif'
    elif polaritas >= 0.5 and subjektivitas > 0.5:
        return 'Pujian Subjektif'
    else:
        return 'Campuran / Ambigu'

# Terapkan ke DataFrame
gabungan_df['kategori_gabungan'] = gabungan_df.apply(
    lambda row: klasifikasi_gabungan(row['polarity_stemmed'], row['subjectivity_stemmed']),
    axis=1
)

# Hitung jumlah tiap kategori gabungan
gabungan_counts = gabungan_df['kategori_gabungan'].value_counts().reindex([
    'Kritik Objektif', 'Kritik Subjektif',
    'Netral Objektif', 'Netral Subjektif',
    'Pujian Objektif', 'Pujian Subjektif',
    'Campuran / Ambigu'
])

# Buat bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=gabungan_counts.index, y=gabungan_counts.values, palette='Spectral')
plt.title('Jumlah Artikel Berdasarkan Kategori Gabungan Polarity dan Subjectivity', fontsize=14)
plt.xlabel('Kategori Gabungan', fontsize=12)
plt.ylabel('Jumlah Artikel', fontsize=12)
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
